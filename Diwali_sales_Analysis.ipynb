## Exploratory data analysis

```python
df.columns
```

```python
sns.countplot(x="Gender" ,data=df , palette='Set2') 
# x axis is denoted by x which will show he gender and in data we have provided dataframs ie df
```

```python
# to add values to the columns 

ax = sns.countplot(x="Gender" ,data=df , palette='viridis') 

for bars in ax.containers:
    ax.bar_label(bars)

    # this is how we can add values of the columns 
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# the same groupby as sql here we are grouping the dataframe on the basis of genders and creating a column named amount which will provide the sum of amount grouped by gender and at last sorting them 
```

```python
sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Gender' , y='Amount' , data=sales_gen , palette='Set2')
```

* from the above graphs we can see that most of the buyers are female and also the purchases are more in the case of females than males.

### Age

```python
sns.countplot(data=df , x = "Age Group" , hue='Gender')
# this hue is used to classify the population on the basis of Gender as specified
```

```python
ax2 = sns.countplot(data=df , x = "Age Group" , hue='Gender' , palette='Set2')

for bars in ax2.containers:
    ax2.bar_label(bars)

# this shows the count of the people on the basis of age and gender who went shopping
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Age Group', y='Amount', data=sales_age , palette='viridis')

# and this plot shows the amount spent by the various age groups
```

* the above graphs we can see that most of the buyers are in the age of 26-35 and are women

### State

```python
df.columns
```

```python
sales_state = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

# this will show the top 10 states which have the most orders

sns.barplot(data=sales_state , x='State' , y='Orders' , palette='Set2')
sns.set(rc={'figure.figsize':(17,5)}) # this is to set the size of the chart
```

```python
sales_state_amt = df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.barplot(data=sales_state_amt , x='State' , y='Amount' , palette='viridis')
sns.set(rc={'figure.figsize':(17,5)})
```

* from the above graphs we can see that the state which have most sales is the uttar pradesh and also this state has spent the most amount , also we have the top 10 states listed in the graphs

### Marital Status

```python
df.columns
```

```python
ax3 = sns.countplot(data=df , x='Marital_Status' , palette="Set2")

for bars in ax3.containers:
    ax3.bar_label(bars)
    
sns.set(rc={'figure.figsize':(7,5)})
```

```python
sales_marital = df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount' , ascending=False)

sns.set(rc={'figure.figsize':(7,5)})

sns.barplot(data = sales_marital , x='Marital_Status' , y='Amount' , hue='Gender' , palette='Set2')
```

* Here we can say by the above graphs that the married women are the most to buy and spend more amount than unmarried

### Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})

ax4 = sns.countplot(data=df , x='Occupation' , palette='viridis')

for bars in ax4.containers:
    ax4.bar_label(bars)
```

```python
sales_occu = df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount' , ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_occu , x='Occupation' , y='Amount' , palette='Set2')
```

* this graph concludes that the most amount spent by any sector is IT sector

```python
ax5 = sns.countplot(data=df , x='Product_Category',palette="viridis")

sns.set(rc={'figure.figsize':(25,5)})

for bars in ax5.containers:
    ax5.bar_label(bars)
```

```python
sales_Product = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount' , ascending=False).head(10)

sns.barplot(data=sales_Product , x='Product_Category' , y='Amount' , palette='viridis')

sns.set(rc={'figure.figsize':(25,5)})
```

* from the above two graphs we came to know that the most selling product was of category clothing but the most amount spent on the category is food

```python
sales_PID = df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders' , ascending=False).head(10)

sns.barplot(data=sales_PID , x='Product_ID' , y='Orders' , palette='magma')
sns.set(rc={'figure.figsize':(15,5)})
```
